```r
airline=read.csv('airlines.csv')
airports=read.csv('airports.csv')
flights_full=read.csv('flights.csv')
```

```r
library(dplyr)
library(tidyr)
library(ggplot2)
library(lubridate)
library(RColorBrewer)
```

```r
f=flights_full 
sapply(f, function(x) sum(is.na(x))) #checking for Nan values in each column
```

```r
f= f %>% filter((DIVERTED!=1)|(CANCELLED!=1))
nrow(f)
```

```r
col_names <- names(factor)
f[col_names] <- lapply(f[col_names] , factor) #converting categorical variables to type factor
```

```r
f <- f %>% 
  mutate(result = if_else(ARRIVAL_DELAY > 15, 1, 0))  #creating a new column result to indicate whether the flight is delayed
f$result <- as.factor(f$result)
flights=f
```

```r
flight=f
flight<- subset(flight, select = c(MONTH, DAY, DAY_OF_WEEK,AIRLINE,FLIGHT_NUMBER,TAIL_NUMBER,ORIGIN_AIRPORT,DESTINATION_AIRPORT,DEPARTURE_DELAY,DISTANCE,ARRIVAL_DELAY))
nrow(flight)
```

```r
flights<- subset(flights, select = -c(TAXI_OUT,TAXI_IN,WHEELS_ON,WHEELS_OFF,YEAR,AIR_SYSTEM_DELAY,SECURITY_DELAY,AIRLINE_DELAY,LATE_AIRCRAFT_DELAY,
                              WEATHER_DELAY, DIVERTED, CANCELLED, CANCELLATION_REASON,
                              FLIGHT_NUMBER, TAIL_NUMBER, AIR_TIME))
nrow(flights)
```

```r
flights=flights %>% drop_na()  #Dropping rows with Nan values
sapply(flights, function(x) sum(is.na(x)))
```

```r
flight <- flight %>% 
  mutate(result = if_else(ARRIVAL_DELAY > 15, 1, 0))
flight$result <- as.factor(flight$result)
```

```r
flights %>% group_by(AIRLINE,MONTH) %>% 
            summarize(avg_delay=mean(ARRIVAL_DELAY)) %>%
            ggplot(aes(AIRLINE,avg_delay,colour=factor(MONTH))) + geom_point(alpha=0.7)+theme_minimal()
```

#### <font color='blue'> [A] Above visualization depicts the average delay per airline across the year(per month). It shows the range of delays per airline. Few airlines have similar average delays across the months, meaning less variance in the delays such as AS,DL,US and VX. Some airlines have a high variance in average delays such as NK,F9,B6 and UA. We can also make out from the colors that the average delay seems to be lesser towards the end of the year. </font>

```r
flights %>% group_by(AIRLINE,result) %>% filter(result==1) %>% tally() %>% arrange(desc(n)) %>% ggplot(aes(x=reorder(AIRLINE,-n),weight=n))+geom_bar(fill="lightblue")+ ylab("No. of Delays")+ xlab("Airline")+theme_minimal()
```

```r
f_air_1=flights %>% group_by(AIRLINE,result) %>% tally() %>% filter(result==1) 
f_air_all=flights %>% group_by(AIRLINE) %>% filter(result==1|result==0) %>% tally()
f_p=f_air_all['AIRLINE']
f_p['percentage']=(f_air_1['n']/f_air_all['n'])*100
f_p %>% arrange(desc(percentage)) %>% ggplot(aes(x=reorder(AIRLINE,-percentage),weight=percentage))+ geom_bar(fill="lightblue") + ylab("Percentage Delay(No. of Delays/Total no. of flights)")+ xlab("Airline")+ theme_minimal()
```

Perform any additional data carpentry and begin filtering results/data and then build, validate, and describe your model(s).

```r
install.packages('lattice')
```

```r
top15 = list('ATL','BOS','CLT','DEN','DFW','DTW','IAH','LAS','LAX','MCO','MSP','ORD','PHX','SEA','SFO')
df_ml <- flights %>% filter((ORIGIN_AIRPORT %in% top15)&(MONTH==1))
unique(df_ml$ORIGIN_AIRPORT)
```

```r
frmla1 <- ARRIVAL_DELAY ~ DEPARTURE_DELAY
```

```r
dt = sort(sample(nrow(df_ml),nrow(df_ml)*0.7))
train_df <- df_ml[dt,]
test_df <- df_ml[-dt,]
```

```r
m1 <- train(frmla1,train_df,method="lm")
summary(m1)
```

```r
intcpt <- coef(m1$finalModel)[1]
intcpt
slp <-coef(m1$finalModel)[2]
slp
```

```r
pred <- predict(m1,test_df)
```

```r
head(cbind(fitted.values(m1),pred))
```

#### <font color='blue'> Above is a Linear regression model which uses departure delay to predict the value of arrival delay (in minutes) for data related to the origin airports belonging to the top 15 busiest airports. The plot for the regression line is shown in the results.</font>

```r
frmla3= result ~ .
```

```r
library(e1071)
m<- naiveBayes(frmla3,data=train_df)
```

```r
pred3=predict(m,test_df)
```

```r
tab3<-table(test_df$result,pred3)
tab3
```

```r
1-(sum(tab3[row(tab3)!=col(tab3)]/sum(tab3)))
```

#### <font color='blue'> Above is the Naive Bayes model used for binary classification. The goal is to predict whether a flight will be delayed(1) or not(0). The model has an accuracy of 94.83% </font>
